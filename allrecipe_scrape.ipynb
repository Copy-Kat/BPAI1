```python
link = []
ingredients = []

with open('output.txt', 'r') as f:
    lines = f.readlines()
    for idx, line in enumerate(lines):
        if idx % 2 == 0 and 'recipedb' in line:
            link.append(line.strip())
        elif 'recipedb' not in line and idx % 2 == 1:
            ingredients.append(line.strip())
        else:
            print(idx, line)
            break
```
