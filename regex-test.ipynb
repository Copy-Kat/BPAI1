```python
import nltk
gram = ("NP: {<VB.*|JJ>?<N.*>*}")

```

```python
extract = []
diff = []
loose_count,strict_count, total = 0, 0, 0

for row in tqdm(df_INQ_agged.itertuples()):
    
    ID = row.ID
    
    ingredients = df_ingredients_agged[df_ingredients_agged['ID'] == ID]['Ingredients'].to_list()[0] 
    
    input_texts = row.INQs
    
    for text in input_texts:
        
        result = test_regex.match(text)
    
        if result is not None and result.groups()[-2] in units:
            
            if result.groups()[-1] != '':

                result = result.groups()[-1]
            
            else:
                
                result = text
                
        else:
                
            result = text

        chunking = nltk.RegexpParser(gram)
        sent_token = nltk.word_tokenize(result)
        tagging = nltk.pos_tag(sent_token)
        tree = chunking.parse(tagging)
        
        result = []

        for subtree in tree.subtrees():
            if subtree.label() == 'NP':
                result.extend(subtree.leaves())
                
        final = " ".join([x[0] for x in result])
        
        # if final is not None:
        #     if final in ingredients:
        #         strict_count += 1
        #         loose_count += 1
        #     else:
        #         for inr in ingredients:
        #             if final in inr or inr in final:
        #                 loose_count += 1
        #                 continue
        #         diff.append(f'{text}, {final}, {ingredients} \n')
        #     extract.append(final)
        
        if final is not None:
            if loose_eval(final, ingredients):
                continue
                
            diff.append(f'{text}, {final}, {ingredients} \n')
            
        total += 1
        
    # if total > 1:
    #     break
        
file = open('compare.txt', 'w')

file.writelines(diff)

file.close()
            
print(loose_count)
print(strict_count)
print(total)
print(loose_count/total)
print(strict_count/total)
```

```python
def loose_eval(s, l):
    for i in l:
        if s in i or i in s:
            return True
    return False
```
